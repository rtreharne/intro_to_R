```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
liverpool = np.random.normal(loc=167.0, scale=20.0, size=1000).round(1)
everton = np.random.normal(loc=170, scale = 15.0, size =1000).round(1)
both = np.append(liverpool, everton)
```

```python
liverpool
```

```python
df_liverpool = pd.DataFrame(liverpool, columns=['Height_cm'])
df_liverpool['Team'] = 'Liverpool'
df_everton = pd.DataFrame(everton, columns=['Height_cm'])
df_everton['Team'] = 'Everton'
```

```python
df_liverpool.head()
```

```python
df_all = pd.concat([df_liverpool, df_everton])
```

```python
df_all.to_csv("fan_data.csv", index=None)
```
